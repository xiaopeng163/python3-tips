```python
def sum1(x, *args):
    return x + sum(args)
sum1(1, 2, 3)
```

```python
def sum2(x, **kwargs):
    print(x, kwargs)

sum2(x=1, a=2,b=3,c=4)
```

```python
def sum3(x, y):
    return x + y

sum3(**{'x': 1, 'y': 2})
```

```python
sum3(x=1,y=2)
```

```python
sum3(**{'x': 1, 'y': 2, 'z': 3})
```
